# COVID-19 Case Visualization & Analysis

Dataset Source: https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide

```python
import pandas as pd
import numpy as np
from collections import defaultdict
import datetime
import plotly.graph_objects as go
```

```python
df = pd.DataFrame.from_csv('CSV_4_COMS-Table 1.csv')
```

```python
df.head()
```

```python
start_date = min(df.index)
end_date = max(df.index)
countries = sorted(np.unique(df['CountryExp']))
```

```python
cases = defaultdict(dict)

for index, row in df.iterrows():
    cases[index][row['CountryExp']] = row['NewConfCases']
```

```python
date = start_date
delta = datetime.timedelta(days=1)

X = []
ys = defaultdict(list)

cum_cases = defaultdict(int)

while (date < end_date):
    X.append(date)
    for country in countries:
        new_cases = cases[date].get(country, 0)
        cum_cases[country] += new_cases
        ys[country].append(cum_cases[country])
    date += delta
    
ys = zip(countries, list(ys.values()))
```

```python
data = [go.Scatter(name=y[0], x=X, y=y[1]) for y in ys]
fig = go.Figure(data=data)
# Use datetime objects to set xaxis range
fig.show()
```
